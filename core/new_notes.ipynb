```python
from pandas import read_csv
import numpy as np
from matplotlib import pyplot as plt
```

```python
source = read_csv("data/test_data.csv", index_col=0, parse_dates=True)
dead = read_csv("dead_reckoning.csv", index_col=0, parse_dates=True)
ins = read_csv("ins.csv", index_col=0, parse_dates=True)
ins['latitude'] = np.rad2deg(ins['latitude'])
ins['longitude'] = np.rad2deg(ins['longitude'])
```

```python
plt.figure(figsize=(10, 10))
plt.plot(source['longitude'], source['latitude'], label='Source', color='blue')
plt.plot(dead['longitude'], dead['latitude'], label='Dead Reckoning', color='red')
plt.plot(ins['longitude'], ins['latitude'], label='INS', color='green')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Comparison of Source, Dead Reckoning, and INS Data')
plt.legend()
plt.grid()
```

```python
lon_error = (source['longitude'] - ins['latitude']) * 60 / 1852
lat_error = (source['latitude']  - ins['longitude']) * 60 / 1852
error = np.sqrt(lon_error**2 + lat_error**2)
altitude_error = (source['altitude'] - ins['altitude'])
```

```python
#plt.plot(lon_error, label='Longitude Error', color='blue')
#plt.plot(lat_error, label='Latitude Error', color='orange')
plt.plot(error, label='Total Error', color='purple')
plt.plot(altitude_error, label='Altitude Error', color='black')
plt.xlabel('Time')
plt.ylabel('Error (meters)')
plt.title('Error Analysis between Source and INS Data')
plt.legend()
#plt.ylim(-0.5e-5, 1e-5)
```
